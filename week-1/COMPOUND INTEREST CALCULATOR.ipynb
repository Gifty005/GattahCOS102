```python
print ("COMPOUND INTEREST CALCULATOR")
P = float(input("Enter principal amount(p):"))
R = float(input("Enter annual interest rate(R%):"))
n = float(input("Enter number of times interest compounded per year(n):"))
t = float(input("Enter number of years (t):"))
 #covert R to percentage         
r = R/100
A = P * (1 + r/n)**(n*t)
CI =A - P
print ("final amount(A)=", A)
print ("compound interest =", CI)
```
