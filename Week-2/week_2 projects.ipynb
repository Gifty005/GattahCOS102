```python
import numpy as np #Import the toolset

def find_roots ():
    # 1. DATA COLLECTION
   #ask the user for the type of equation
  degree = int(input("Enter equation degree (2,3,4): " ))
    
  #2.BUILDING THE ARRAY
  #We create an empty list to store the 'A,B,C....' values.
  coeffs = []
  #we loop to get exactly the number of coefficients needed.
  #Aquadratic (degree 2) needs 3 coefficients (A,B,C).
  for i in range(degree + 1):
      val = float(input(f"Enter coefficient for term {i}: "))
      coeffs.append(val)
   #3.PROCESSING (The 'Black box')
  #np.roots finds where the graph hits the x_axis.
  results = np.roots(coeffs)
  #4.OUTPUT FORMATTING
  print ("\nCalculated Roots: ")
  for root in results:
  #complex numbers in python use 'j' instead of 'i'.
  #we round to two decimal places to keep it clean.
   print(f"->{np.round(root,2)}")
    
find_roots()
```
