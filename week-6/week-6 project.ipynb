```python
import tkinter as tk
import random

class Employee():
    def __init__(self):
        self.employees = ["Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", 
                         "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James", 
                         "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso", 
                         "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"]
        self.tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]
        self.attendance = {}
    
    def check(self, name):
        return name in self.employees
    
    def mark(self, name):
        if name not in self.attendance:
            self.attendance[name] = True
            return True
        return False
    
    def task(self, name):
        return random.choice(self.tasks)
    
    def deny(self, name):
        return name + " is not an employee"

# Create window
root = tk.Tk()
root.title("Employee System")
system = Employee()

# Input
tk.Label(root, text="Name:").pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Output area
output = tk.Text(root, height=10)
output.pack()

# Button functions
def check():
    n = name_entry.get()
    if system.check(n):
        output.insert(tk.END, "YES - " + n + " works here\n")
    else:
        output.insert(tk.END, system.deny(n) + "\n")

def attend():
    n = name_entry.get()
    if system.check(n):
        if system.mark(n):
            output.insert(tk.END, n + " marked present\n")
        else:
            output.insert(tk.END, n + " already marked\n")
    else:
        output.insert(tk.END, system.deny(n) + "\n")

def assign():
    n = name_entry.get()
    if system.check(n):
        output.insert(tk.END, n + " -> " + system.task(n) + "\n")
    else:
        output.insert(tk.END, system.deny(n) + "\n")

# Buttons
tk.Button(root, text="Check", command=check, bg="green", fg="white").pack()
tk.Button(root, text="Attendance", command=attend, bg="blue", fg="white").pack()
tk.Button(root, text="Task", command=assign, bg="orange", fg="white").pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

class DeliveryCalculator():
    def __init__(self):
        # Prices for PAU
        self.pau_heavy = 2000   # 10kg or more
        self.pau_light = 1500   # less than 10kg
        
        # Prices for Epe
        self.epe_heavy = 5000   # 10kg or more
        self.epe_light = 4000   # less than 10kg
    
    def calculate(self, location, weight):
        if location == "PAU":
            if weight >= 10:
                return self.pau_heavy
            else:
                return self.pau_light
        else:  # Epe
            if weight >= 10:
                return self.epe_heavy
            else:
                return self.epe_light

# Create window
root = tk.Tk()
root.title("Delivery Cost")
root.geometry("400x450")

# Create calculator
calc = DeliveryCalculator()

# Title
tk.Label(root, text="DELIVERY COST CALCULATOR", font=("Arial", 14, "bold")).pack(pady=10)

# Location selection
tk.Label(root, text="Where to deliver?").pack()
location_var = tk.StringVar(value="PAU")
tk.Radiobutton(root, text="PAU", variable=location_var, value="PAU").pack()
tk.Radiobutton(root, text="Epe", variable=location_var, value="Epe").pack()

# Weight input
tk.Label(root, text="Weight in kg:").pack(pady=10)
weight_entry = tk.Entry(root)
weight_entry.pack()

# Result label
result_label = tk.Label(root, text="", font=("Arial", 12, "bold"))
result_label.pack(pady=20)

# Calculate function
def calculate_price():
    try:
        weight = float(weight_entry.get())
        location = location_var.get()
        price = calc.calculate(location, weight)
        
        if weight >= 10:
            result_label.config(text=f"Cost: N{price}", fg="green")
        else:
            result_label.config(text=f"Cost: N{price}", fg="blue")
    except:
        messagebox.showerror("Error", "Enter a number!")

# Calculate button
tk.Button(root, text="CALCULATE", command=calculate_price, bg="red", fg="white").pack(pady=10)

# Info box
info = tk.Text(root, height=8, width=35)
info.insert(tk.END, "PRICES:\n")
info.insert(tk.END, "PAU: 10kg+ = N2000\n")
info.insert(tk.END, "PAU: less = N1500\n")
info.insert(tk.END, "Epe: 10kg+ = N5000\n")
info.insert(tk.END, "Epe: less = N4000\n")
info.config(state='disabled')
info.pack()

root.mainloop()
```
