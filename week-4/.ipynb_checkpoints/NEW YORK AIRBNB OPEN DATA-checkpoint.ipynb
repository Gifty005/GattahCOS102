```python
import pandas as pd #for handling data like excel in python
import matplotlib.pyplot as plt #for creating charts
import seaborn as sns          #for cresting beautiful charts (built on matplotlib)

df = pd.read_csv('AB_NYC_2019.csv')
df
```

```python
df.shape
```

```python
df.columns
```

```python
print(df['price'].head())
```

```python
print("Total listings:", len(df)) #checking how many rows
```

```python
print("Missing values in price:", df['price'].isnull().sum()) #checking for missing values
```

```python
#Calculating basic statistics
print("Average price:", df['price'].mean())
print("Median price:", df['price'].median())
print("Price range:", df['price'].min(), "-", df['price'].max())
```

```python
print("Average price by borough:")
print(df.groupby('neighbourhood_group')['price'].mean())
```

```python
#Calculate averages
borough_avg = df.groupby('neighbourhood_group')['price'].mean()
#create chart
plt.bar(borough_avg.index, borough_avg.values, color = 'skyblue')
plt.title('Average price by borough', fontsize=14)
plt.xlabel('Borough')
plt.ylabel('Price ($)')
#show it
plt.show()
```

```python
#create a new figure and make it 10 inches wide ,6 inches tall
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50, color='lightgreen', edgecolor='black')
plt.title('Distribution of airbnb prices in NYC', fontsize=14, fontweight='bold')
#add x- axis label
plt.xlabel('Price per Night (USD)', fontsize=12)
plt.ylabel('Number of listings', fontsize=12)
avg_price = df['price'].mean()
median_price = df['price'].median()
print("Average price: $", avg_price)
print("median price: $", median_price)
plt.axvline(avg_price, color='red', linestyle='--', linewidth=2, label='Average')
plt.axvline(median_price, color='blue', linestyle='--',linewidth=2, label ='Median')
plt.legend()
plt.show()
```

```python
print("Room types available:")
print(df['room_type'].unique())#.unique shows all the different values
room_counts=df['room_type'].value_counts()
print("Number of listings by room type:")
print(room_counts)
      
```

```python
#create a new figure
plt.figure(figsize=(10, 6))
#room_count.index = the room types name(what goes on x-axis
#room_count.values = the counts (height of bars)
plt.bar(room_counts.index,room_counts.values, color='skyblue', edgecolor='black')
plt.title('Number of Listings by Room Type', fontsize=14, fontweight='bold')
plt.xlabel('Room Type', fontsize=10)
plt.ylabel('Number of Listings', fontsize=10)

for i,count in enumerate(room_counts.values):
    plt.text(i, count + 300, str(count), ha='center', fontsize=10)
    plt.show()


```
