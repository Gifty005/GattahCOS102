```python
import tkinter as tk
from tkinter import messagebox

class Zenith:
    """Base class for all banking divisions"""
    
    def __init__(self, division_name):
        self.division_name = division_name
    
    def mutual_services(self):
        """Mutual services common to all divisions"""
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance"
        ]
    
    def unique_services(self):
        """Will be overridden by child classes"""
        pass

class RetailBanking(Zenith):
    def __init__(self):
        super().__init__("Retail Banking")
    
    def unique_services(self):
        return [ 
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"
        ]
    
    def mutual_services(self):
        """Override mutual services for Retail division"""
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance",
            "Retirement and education accounts",  # Also unique to Retail
            "Loans and mortgages",
            "Checking and saving"
        ]

class GlobalBanking(Zenith):
    def __init__(self):
        super().__init__("Global Banking")
    
    def unique_services(self):
        return [
            "Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]

class CommercialBanking(Zenith):
    def __init__(self):
        super().__init__("Commercial Banking")
    
    def unique_services(self):
        return [
            "Advisory services"
        ]
    
    def mutual_services(self):
        """Override mutual services for Commercial division"""
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance",
            "Advisory services"  # This is unique to Commercial but treated as mutual here
        ]

def show_services():
    """Display services based on selected division"""
    name = name_entry.get().strip()
    division = division_var.get().lower()
    
    # Validate input
    if not name:
        messagebox.showwarning("Input Error", "Please enter employee name")
        return
    
    # Create appropriate object using polymorphism
    if division == "retail":
        obj = RetailBanking()
    elif division == "global":
        obj = GlobalBanking()
    elif division == "commercial":
        obj = CommercialBanking()
    else:
        result_label.config(text="Invalid division selected!")
        return
    
    # Get services (polymorphism in action)
    services = obj.mutual_services()  # Each division can have different mutual services
    
    # Build output string
    output = f"Employee: {name}\n"
    output += f"Division: {obj.division_name}\n"
    output += "=" * 40 + "\n"
    output += "Services Provided:\n\n"
    
    for service in services:
        output += "• " + service + "\n"
    
    # Display in label
    result_label.config(text=output)

# Create main window
root = tk.Tk()
root.title("Zenith Bank Nigeria - Division Services")
root.geometry("500x450")
root.configure(bg='light gray')

# Title
title_label = tk.Label(root, text="Zenith Bank Service Management", 
                       font=("Arial", 14, "bold"), bg='light gray')
title_label.pack(pady=10)

# Employee name input
tk.Label(root, text="Employee Name:", font=("Arial", 11), 
         bg='light gray').pack(pady=5)
name_entry = tk.Entry(root, width=35, font=("Arial", 10))
name_entry.pack(pady=5)

# Division selection using dropdown (better than text entry)
tk.Label(root, text="Select Division:", font=("Arial", 11), 
         bg='light gray').pack(pady=5)
division_var = tk.StringVar(value="retail")
division_menu = tk.OptionMenu(root, division_var, "retail", "global", "commercial")
division_menu.config(width=20, font=("Arial", 10))
division_menu.pack(pady=5)

# Show services button
show_button = tk.Button(root, text="Show Services", command=show_services,
                        bg="blue", fg="white", font=("Arial", 11), 
                        width=20, height=1)
show_button.pack(pady=15)

# Result display
result_label = tk.Label(root, text="", font=("Arial", 10), 
                        justify=tk.LEFT, bg='white', 
                        relief=tk.SUNKEN, padx=10, pady=10)
result_label.pack(pady=10, padx=20, fill=tk.BOTH, expand=True)

# Start the GUI
root.mainloop()
```
